```python
def load_parallel_data(en_path, vi_path, limit=None):
    with open(en_path, encoding='utf-8') as f_en, open(vi_path, encoding='utf-8') as f_vi:
        en_lines = f_en.read().strip().split('\n')
        vi_lines = f_vi.read().strip().split('\n')
        pairs = [(v,e) for v,e in zip(vi_lines, en_lines) if v and e]
        if limit:
            pairs = pairs[:limit]
        
        return pairs
```
